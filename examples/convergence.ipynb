*On the convergence of physics informed neural networks for linear second-order elliptic and parabolic type PDEs*

Общая ошибка ри обучении нейронных сетей делится на три компоненты: 1) ошибка аппроксимации, 2) ошибка оптимизации, 3) обобщающая ошибка.

1) Ошибка апроксимации активно изучается и есть теоремы об универсальной аппроксимации
2) Ошибка оптимизации менее изучена, так как оптимизируемые функции невыпуклые. Также используется много эвристик и инженерных подходв к оптимизации. Нет никакой гарантии, что градиентная оптимизация найдет глобальный минимум для общих задач машинного обучения, включая PINN. Это одна из открытых проблем.
3) Ошибка обобщения возникает из-за использования конечного набор узлов при обучении. В машинном обучении, ошибка обобщения относится к мере точности прогноза
на данных, которые модель не видела.



Статья посвящена верхней оценке ошибки аппроксимации.
